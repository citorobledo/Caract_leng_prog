### Buscar elemento en una colección
- Imperativo
    - COBOL sigue el paradigma imperativo, indicando paso a paso cómo recorrer la colección.

```python
IDENTIFICATION DIVISION.
PROGRAM-ID. BUSCAR-ELEMENTO.
DATA DIVISION.
WORKING-STORAGE SECTION.
01 VALOR-BUSCADO    PIC 9(2) VALUE 25.
01 ENCONTRADO         PIC X    VALUE 'N'.
01 I                		    PIC 9(2).

01 ARREGLO-INICIAL.
   05 FILLER PIC 9(2) VALUE 10.
   05 FILLER PIC 9(2) VALUE 20.
   05 FILLER PIC 9(2) VALUE 25.
   05 FILLER PIC 9(2) VALUE 30.
   05 FILLER PIC 9(2) VALUE 40.

01 ARREGLO REDEFINES ARREGLO-INICIAL.
   05 ELEMENTO PIC 9(2) OCCURS 5 TIMES.

PROCEDURE DIVISION.
    PERFORM VARYING I FROM 1 BY 1
            UNTIL I > 5 OR ENCONTRADO = 'S'
        IF ELEMENTO(I) = VALOR-BUSCADO
            MOVE 'S' TO ENCONTRADO
            DISPLAY "Encontrado en posicion " I
        END-IF
    END-PERFORM.

    IF ENCONTRADO = 'N'
        DISPLAY "No encontrado."
    END-IF.

    STOP RUN.
```

- Orientado a Objetos
    - Python:  Se encapsula la lógica en una clase, con métodos que operan sobre los datos.

```python
class Buscador:
    def __init__(self, coleccion):
        self.coleccion = coleccion

    def buscar(self, valor):
        for i, elemento in enumerate(self.coleccion):
            if elemento == valor:
                return f"Encontrado en posición { 1+i }"
        return "No encontrado"

# Ejemplo de uso
b = Buscador([10, 20, 25, 30, 40])
print(b.buscar(25))
```

- Funcional
  - Haskell: Usa funciones puras y estructuras inmutables; no hay bucles ni estados mutables.

```python

-- Implementacion manual con recursion
buscarElemento :: (Eq a) => a -> [a] -> Maybe Int
buscarElemento _ []     = Nothing
buscarElemento buscado (x:xs)
    | buscado == x = Just 0
    | otherwise    = fmap (+1) (buscarElemento buscado xs)

-- Programa principal
main :: IO ()
main = do
    let lista = [10, 20, 25, 30, 40]

    case buscarElemento 25 lista of
        Just i  -> putStrLn ("Encontrado en posicion " ++ show i)
        Nothing -> putStrLn "No encontrado."

    case buscarElemento 99 lista of
        Just i  -> putStrLn ("Encontrado en posicion " ++ show i)
        Nothing -> putStrLn "No encontrado."
```

- Lógico
  - Prolog: En Prolog se declaran hechos y reglas; el motor lógico infiere la posición del elemento.

```python
buscar_elemento(X, [X|_], 0).
buscar_elemento(X, [_|T], N) :-
    buscar_elemento(X, T, N1),
    N is N1 + 1.

% Ejemplo de consulta:
%?- buscar_elemento(25, [10,20,25,30,40], Pos).
% Pos = 2.


/** <examples>
?- buscar_elemento(25, [10,20,25,30,40], Pos).
*/
```

### Ordenar elementos en una colección


- Imperativo
  - COBOL: ordenamiento. 

```python
Imperativo — COBOL: ordenamiento. 
      
IDENTIFICATION DIVISION.
PROGRAM-ID. ORDENAR-ELEMENTOS.
DATA DIVISION.
WORKING-STORAGE SECTION.

01 ARREGLO-INICIAL.
   05 FILLER PIC 9(2) VALUE 40.
   05 FILLER PIC 9(2) VALUE 10.
   05 FILLER PIC 9(2) VALUE 55.
   05 FILLER PIC 9(2) VALUE 20.
   05 FILLER PIC 9(2) VALUE 25.
   05 FILLER PIC 9(2) VALUE 30.

01 ARREGLO REDEFINES ARREGLO-INICIAL.
   05 ELEMENTO PIC 9(2) OCCURS 6 TIMES.

01 I    PIC 9(2).
01 J    PIC 9(2).
01 TEMP PIC 9(2).

PROCEDURE DIVISION.
    PERFORM VARYING I FROM 1 BY 1 UNTIL I > 5
        COMPUTE J = I + 1
        PERFORM UNTIL J > 6
            IF ELEMENTO(I) > ELEMENTO(J)
                MOVE ELEMENTO(I) TO TEMP
                MOVE ELEMENTO(J) TO ELEMENTO(I)
                MOVE TEMP        TO ELEMENTO(J)
            END-IF
            ADD 1 TO J
        END-PERFORM
    END-PERFORM.

    DISPLAY "Arreglo ordenado:".
    PERFORM VARYING I FROM 1 BY 1 UNTIL I > 6
        DISPLAY ELEMENTO(I)
    END-PERFORM.

    STOP RUN.
```

- Orientado a Objetos
  - Python: Encapsula la lógica en una clase y usa el método sort()

```python

class Ordenador:
    def __init__(self, coleccion):
        self.coleccion = coleccion

    def ordenar(self):
        self.coleccion.sort()
        return self.coleccion

# Ejemplo de uso
o = Ordenador([40, 10, 30, 20, 25])
print(o.ordenar())
```

- Funcional
  - Haskell: Usa recursión y listas por comprensión para implementar quicksort. 

```python

quicksort :: (Ord a) => [a] -> [a]
quicksort []     = []
quicksort (p:xs) =
    quicksort menores ++ [p] ++ quicksort mayores
  where
    menores = [ x | x <- xs, x <= p ]
    mayores = [ x | x <- xs, x >  p ]

main :: IO ()
main = do
    let lista = [40, 10, 30, 20, 25]
    print (quicksort lista)

salida: [10,20,25,30,40]
```

### Gestionar una lista de tareas pendientes

- Imperativo
  - COBOL: Se gestiona con registros y banderas de completado.

```python
 
IDENTIFICATION DIVISION.
PROGRAM-ID. GESTIONAR-TAREAS.

DATA DIVISION.
WORKING-STORAGE SECTION.

01 CANT-TAREAS      PIC 9(2) VALUE 0.
01 CONT-PENDIENTES  PIC 9(2) VALUE 0.
01 I                PIC 9(2).
01 PRIO             PIC 9(1).
01 OPCION           PIC 9(1) VALUE 0.
01 NUEVO-TITULO     PIC X(30).
01 NUEVA-PRIORIDAD  PIC 9(1).
01 NUM-TAREA        PIC 9(2).

01 LISTA-TAREAS.
   05 TAREA OCCURS 10 TIMES.
      10 TITULO    PIC X(30).
      10 PRIORIDAD PIC 9(1).
      10 ESTADO    PIC X(1).

PROCEDURE DIVISION.

    PERFORM UNTIL OPCION = 5
        DISPLAY " =============================="
        DISPLAY " =      GESTION DE TAREAS     ="
        DISPLAY " =============================="
        DISPLAY "1. Agregar tarea"
        DISPLAY "2. Marcar tarea como completa"
        DISPLAY "3. Ver lista de tareas"
        DISPLAY "4. Ver pendientes por prioridad"
        DISPLAY "5. Salir"
        DISPLAY "Opcion: "
        ACCEPT OPCION

        EVALUATE OPCION
            WHEN 1 PERFORM AGREGAR-TAREA
            WHEN 2 PERFORM MARCAR-COMPLETA
            WHEN 3 PERFORM VER-LISTA
            WHEN 4 PERFORM MOSTRAR-PENDIENTES
            WHEN 5 DISPLAY "Hasta luego."
            WHEN OTHER DISPLAY "Opcion invalida. Ingrese del 1 al 5."
        END-EVALUATE
    END-PERFORM.

    STOP RUN.

AGREGAR-TAREA.
    IF CANT-TAREAS >= 10
        DISPLAY "Lista llena. No se pueden agregar mas tareas."
    ELSE
        DISPLAY "--- AGREGAR TAREA ---"
        DISPLAY "Titulo (max 30 caracteres): "
        ACCEPT NUEVO-TITULO
        DISPLAY "Prioridad (1=Alta  2=Media  3=Baja): "
        ACCEPT NUEVA-PRIORIDAD
        IF NUEVA-PRIORIDAD < 1 OR NUEVA-PRIORIDAD > 3
            DISPLAY "Prioridad invalida. Ingrese 1, 2 o 3."
        ELSE
            ADD 1 TO CANT-TAREAS
            MOVE NUEVO-TITULO    TO TITULO(CANT-TAREAS)
            MOVE NUEVA-PRIORIDAD TO PRIORIDAD(CANT-TAREAS)
            MOVE 'P'             TO ESTADO(CANT-TAREAS)
            DISPLAY "Tarea agregada correctamente."
        END-IF
    END-IF.

MARCAR-COMPLETA.
    IF CANT-TAREAS = 0
        DISPLAY "No hay tareas registradas."
    ELSE
        PERFORM VER-LISTA
        DISPLAY "Numero de tarea a marcar como completa: "
        ACCEPT NUM-TAREA
        IF NUM-TAREA < 1 OR NUM-TAREA > CANT-TAREAS
            DISPLAY "Numero de tarea invalido."
        ELSE
            IF ESTADO(NUM-TAREA) = 'C'
                DISPLAY "Esa tarea ya estaba marcada como completa."
            ELSE
                MOVE 'C' TO ESTADO(NUM-TAREA)
                DISPLAY "Tarea marcada como completa."
            END-IF
        END-IF
    END-IF.

VER-LISTA.
    IF CANT-TAREAS = 0
        DISPLAY "No hay tareas registradas."
    ELSE
        DISPLAY " =============================="
        DISPLAY " =     LISTA DE TAREAS        ="
        DISPLAY " =============================="
        PERFORM VARYING I FROM 1 BY 1 UNTIL I > CANT-TAREAS
            DISPLAY I ") " TITULO(I)
            EVALUATE PRIORIDAD(I)
                WHEN 1 DISPLAY "   Prioridad : ALTA"
                WHEN 2 DISPLAY "   Prioridad : MEDIA"
                WHEN 3 DISPLAY "   Prioridad : BAJA"
            END-EVALUATE
            IF ESTADO(I) = 'C'
                DISPLAY "   Estado    : COMPLETA"
            ELSE
                DISPLAY "   Estado    : PENDIENTE"
            END-IF
        END-PERFORM
    END-IF.

MOSTRAR-PENDIENTES.
    MOVE 0 TO CONT-PENDIENTES
    PERFORM VARYING I FROM 1 BY 1 UNTIL I > CANT-TAREAS
        IF ESTADO(I) = 'P'
            ADD 1 TO CONT-PENDIENTES
        END-IF
    END-PERFORM.

    IF CONT-PENDIENTES = 0
        DISPLAY "No hay tareas pendientes."
    ELSE
        DISPLAY " =============================="
        DISPLAY " =      TAREAS PENDIENTES     ="
        DISPLAY " = (ordenadas por prioridad)  ="
        DISPLAY " =============================="
        PERFORM VARYING PRIO FROM 1 BY 1 UNTIL PRIO > 3
            EVALUATE PRIO
                WHEN 1 DISPLAY "--- PRIORIDAD ALTA ---"
                WHEN 2 DISPLAY "--- PRIORIDAD MEDIA ---"
                WHEN 3 DISPLAY "--- PRIORIDAD BAJA ---"
            END-EVALUATE
            PERFORM VARYING I FROM 1 BY 1 UNTIL I > CANT-TAREAS
                IF ESTADO(I) = 'P' AND PRIORIDAD(I) = PRIO
                    DISPLAY I ") " TITULO(I)
                END-IF
            END-PERFORM
        END-PERFORM
    END-IF.
```

- Orientado a Objetos
  - Python: Encapsula tareas y lista en clases, con métodos para operar. 

```python

class Tarea:
    def __init__(self, titulo, prioridad):
        self.titulo = titulo
        self.prioridad = prioridad
        self.completada = False

    def completar(self):
        self.completada = True

class ToDoList:
    def __init__(self):
        self.tareas = []

    def agregar(self, titulo, prioridad):
        self.tareas.append(Tarea(titulo, prioridad))

    def mostrar_pendientes(self):
        pendientes = [t for t in self.tareas if not t.completada]
        print("Lista de pendientes \n")
        for t in sorted(pendientes, key=lambda x: x.prioridad):
            print(f"{t.titulo} \nPrioridad: {t.prioridad} \n")

# Ejemplo de uso
lista = ToDoList()
lista.agregar("Comprar pan", 2)
lista.agregar("Estudiar Python", 1)
lista.agregar("Pagar servicios", 3)

lista.mostrar_pendientes()
lista.tareas[1].completar()
lista.mostrar_pendientes()
```

- Funcional
  - Haskell: Usa estructuras inmutables y funciones puras para manipular la lista. 


```python

import Data.List (sortBy)
import Data.Ord  (comparing)

-- ==========================================
-- TIPOS DE DATOS
-- ==========================================

data Prioridad = Alta | Media | Baja
    deriving (Show, Eq, Ord)

data Estado = Pendiente | Completa
    deriving (Show, Eq)

data Tarea = Tarea
    { titulo    :: String
    , prioridad :: Prioridad
    , estado    :: Estado
    } deriving (Show)

type ListaTareas = [Tarea]

-- ==========================================
-- OPERACIONES PURAS
-- ==========================================

agregarTarea :: String -> Prioridad -> ListaTareas -> ListaTareas
agregarTarea t p tareas =
    tareas ++ [Tarea { titulo = t, prioridad = p, estado = Pendiente }]

marcarCompleta :: Int -> ListaTareas -> ListaTareas
marcarCompleta n tareas
    | n < 1 || n > length tareas = tareas
    | otherwise =
        [ if j == n - 1 then t { estado = Completa } else t
        | (j, t) <- zip [0..] tareas ]

pendientesOrdenadas :: ListaTareas -> ListaTareas
pendientesOrdenadas tareas =
    sortBy (comparing prioridad)
           (filter (\t -> estado t == Pendiente) tareas)

-- ==========================================
-- PROGRAMA PRINCIPAL
-- ==========================================

main :: IO ()
main = do

    -- Estado inicial: lista vacia
    let lista0 = []

    -- Agregar tareas
    let lista1 = agregarTarea "Estudiar parcial"  Alta  lista0
    let lista2 = agregarTarea "Comprar comida"    Baja  lista1
    let lista3 = agregarTarea "Entregar TP"       Alta  lista2
    let lista4 = agregarTarea "Llamar al medico"  Media lista3
    let lista5 = agregarTarea "Leer capitulo 5"   Media lista4

    putStrLn "=== LISTA INICIAL ==="
    mapM_ print lista5

    -- Marcar tareas como completas
    let lista6 = marcarCompleta 1 lista5  -- "Estudiar parcial"
    let lista7 = marcarCompleta 3 lista6  -- "Entregar TP"

    putStrLn "\n=== DESPUES DE MARCAR COMPLETAS ==="
    mapM_ print lista7

    -- Mostrar pendientes ordenadas por prioridad
    let pendientes = pendientesOrdenadas lista7

    putStrLn "\n=== PENDIENTES POR PRIORIDAD ==="
    mapM_ print pendientes
```
